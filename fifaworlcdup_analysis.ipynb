```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data into a DataFrame
df = pd.read_csv('WorldCupMatches.csv')

# Assuming the CSV file has columns 'Stadium' and 'Attendance'
# 'Stadium' contains the stadium name, 'Attendance' contains the attendance for each match

# Group by stadium and calculate the average attendance
average_attendance_by_stadium = df.groupby('Stadium')['Attendance'].mean().sort_values(ascending=False)

# Plot the top 10 stadiums with the highest average attendance
top_stadiums = average_attendance_by_stadium.head(10)
top_stadiums.plot(kind='bar', figsize=(10, 6), color='skyblue')

# Customize the plot
plt.title('Top 10 Stadiums with Highest Average Attendance')
plt.xlabel('Stadium')
plt.ylabel('Average Attendance')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a DataFrame `df` with columns 'Team' and 'Result'
# Example data
data = {
    'Team': ['Germany', 'Argentina', 'Netherlands', 'Brazil', 'Italy', 'France', 'Germany', 'Brazil', 'France', 'Croatia'],
    'Result': ['Runner-up', 'Runner-up', 'Runner-up', 'Runner-up', 'Runner-up', 'Runner-up', 'Winner', 'Winner', 'Winner', 'Winner']
}

df = pd.DataFrame(data)

# Count the occurrences of each team as runner-up and winner
team_counts = df['Team'].value_counts()
runner_up_counts = df[df['Result'] == 'Runner-up']['Team'].value_counts()
winner_counts = df[df['Result'] == 'Winner']['Team'].value_counts()

# Combine counts into a single DataFrame
results = pd.DataFrame({'Runner-up': runner_up_counts, 'Winner': winner_counts}).fillna(0)
results['Total'] = results.sum(axis=1)

# Sort by total appearances (descending)
results = results.sort_values(by='Total', ascending=False)

# Plotting the data
plt.figure(figsize=(10, 6))

# Plot bars for runner-up and winner counts
results[['Runner-up', 'Winner']].plot(kind='bar', stacked=True, color=['salmon', 'skyblue'])

# Adding labels and title
plt.xlabel('Country')
plt.ylabel('Number of Appearances')
plt.title('World Cup Final Results by Nation (Runner-up vs Winner)')
plt.legend(title='Result')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a DataFrame `df` with columns 'Team' and 'Participations'
# Example data
data = {
    'Team': ['Brazil', 'Germany', 'Italy', 'Argentina', 'France', 'Spain', 'England', 'Netherlands', 'Uruguay', 'Sweden'],
    'Participations': [21, 19, 19, 17, 15, 15, 15, 10, 13, 12]
}

df = pd.DataFrame(data)

# Sort the DataFrame by the number of participations
df_sorted = df.sort_values(by='Participations', ascending=False)

# Plotting the data
plt.figure(figsize=(10, 6))

# Create a bar plot
bars = plt.bar(df_sorted['Team'], df_sorted['Participations'], color='skyblue')

# Adding data labels to the bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, int(yval), va='bottom', ha='center')

# Adding labels and title
plt.xlabel('Team')
plt.ylabel('Number of Participations')
plt.title('Teams with the Most World Cup Participations')

# Rotating x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Displaying the plot
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,15))
matches = matches.rename(columns={'Half-time Home Goals':"first half home goals",
                                  'Half-time Away Goals':"first half away goals"})

matches["second half home goals"] = matches["Home Team Goals"] - matches["first half home goals"]
matches["second half away goals"] = matches["Away Team Goals"] - matches["first half away goals"]

plt.subplot(211)
sns.kdeplot(matches["first half home goals"],color="b",linewidth=2)
sns.kdeplot(matches["second half home goals"],color="r",linewidth=2)
plt.xticks(np.arange(0,9,1))
plt.title("Distribution of first and second Half - Home Team Goals",color='b')

plt.subplot(212)
sns.kdeplot(matches["first half away goals"],color="b",linewidth=2)
sns.kdeplot(matches["second half away goals"],color="r",linewidth=2)
plt.xticks(np.arange(0,9,1))
plt.title("Distribution of first and second Half - Away Team Goals",color='b')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('WorldCupMatches.csv')

# Filter the dataset to include years from 1930 to 2014
df_filtered = df[(df['Year'] >= 1930) & (df['Year'] <= 2014)]

# Plotting the data
plt.figure(figsize=(10, 6))

plt.plot(df_filtered['Year'], df_filtered['Home Team Goals'], label='Home Goals', marker='o')
plt.plot(df_filtered['Year'], df_filtered['Away Team Goals'], label='Away Goals', marker='o')

# Adding labels and title
plt.xlabel('Year')
plt.ylabel('Goals')
plt.title('FIFA World Cup Home and Away Goals by Year (1930-2014)')
plt.legend()

# Displaying the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read data from CSV file
fifa_matches = pd.read_csv('WorldCups.csv')  # Replace 'fifa_matches.csv' with the path to your CSV file

# Filter data for years from 1930 to 2014
fifa_matches = fifa_matches[(fifa_matches['Year'] >= 1930) & (fifa_matches['Year'] <= 2014)]

# Group by year and find the maximum total goals scored
max_goals_per_year = fifa_matches.groupby('Year')['GoalsScored'].max().reset_index()

# Plotting
plt.figure(figsize=(10, 6))

# Plot highest total goals per year
plt.plot(max_goals_per_year['Year'], max_goals_per_year['GoalsScored'], marker='o')

# Adding labels and title
plt.xlabel('Year')
plt.ylabel('Highest Total Goals Scored')
plt.title('Highest Total Goals Scored During a FIFA World Cup Match (1930-2014)')

# Show plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the dataset
df = pd.read_csv('WorldCups.csv')



# Aggregate total matches played per country
matches_per_country = df.groupby('Country')['MatchesPlayed'].sum().reset_index()

# Sort by total matches played
matches_per_country = matches_per_country.sort_values(by='MatchesPlayed', ascending=False)

# Plot
plt.figure(figsize=(10, 6))
plt.bar(matches_per_country['Country'], matches_per_country['MatchesPlayed'])
plt.xticks(rotation=90)
plt.xlabel('Country')
plt.ylabel('Total Matches Played')
plt.title('Total World Cup Matches Played in Each Country')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data: FIFA World Cup years and corresponding coaches
fifa_years = [
    1930, 1934, 1938, 1950, 1954, 1958, 1962, 1966, 1970, 1974, 1978, 1982, 1986, 1990, 1994, 1998, 2002, 2006, 2010, 2014
]

coaches = [
    "Uruguay: Alberto Suppici", "Italy: Vittorio Pozzo", "Italy: Vittorio Pozzo", "Uruguay: Juan López Fontana", "West Germany: Sepp Herberger",
    "Brazil: Vicente Feola", "Brazil: Aymoré Moreira", "England: Alf Ramsey", "Brazil: Mário Zagallo", "West Germany: Helmut Schön",
    "Argentina: César Luis Menotti", "Italy: Enzo Bearzot", "Argentina: Carlos Bilardo", "West Germany: Franz Beckenbauer", "Brazil: Carlos Alberto Parreira",
    "France: Aimé Jacquet", "Brazil: Luiz Felipe Scolari", "Italy: Marcello Lippi", "Spain: Vicente del Bosque", "Germany: Joachim Löw"
]

# Plotting the graph
plt.figure(figsize=(20, 18))
plt.bar(fifa_years, range(len(fifa_years)), color='skyblue')
plt.xticks(fifa_years, rotation=90)
plt.xlabel('FIFA World Cup Year')
plt.ylabel('Coaches')
plt.title('Coaches in FIFA World Cup (1930-2014)')
plt.tight_layout()

# Annotating coach names
for i, coach in enumerate(coaches):
    plt.annotate(coach, (fifa_years[i], i), xytext=(10,-5), textcoords='offset points')

plt.show()
```
